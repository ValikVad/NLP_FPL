### Series

Series - одномерный похожий на массив объект, содержащий массив данных **и ассоциированный с им массив меток**, который называется *индексом*.

Простейший объект Series состоит только из массива данных:

````python
obj = pd.Series([4, 7, -5, 3])
obj
````

 Поскольку индекс был нами не задан - он сгенерировался по умолчанию - целые числа от $0$ до $N-1$ ($N$ - длина массива).

 Атрибуты values и index объекта Series позволяют соответственно получить массив данных и связанный с ним массив меток.

````python
obj.values
````

````python
obj.index  # like range(4)
````

Часто создают Series с индексом, идентифицирующий элемент данных:

````python
obj2.index
````

### Объект DataFrame

Структура данных Data Frame представляет собой "матрицу" (таблицу), в которой все элементы в одном столбце принадлежат одному типу данных. Типы данных самих столбцов могут различаться.
Такая структура отлично подходит (и была создана) для представления данных типа *наблюдения × признаки*: каждая строчка в data frame соотвествует **наблюдению** (**объекту**), а столбцы содержат **значения признаков**, которыми эти объекты характеризуются.

В объекте DataFrame хранятся два индекса:
* по строкам
* по столбцам

Можно считать, что это словарь объектов Series.

Существует много способов сконструировать объект DataFrame, один из самых распространенных - на основе словаря списков одинаковой длины или массивов NumPy:

### Reindexing

Для объектов Pandas критически важен метод ```reindex```, т.е. возможность создания нового объекта, данные в котором согласуются с новым индексом. Простой пример:

````python
obj = pd.Series([4.5, 7.2, -5.3, 3.6], index=['d', 'b', 'a', 'c'])
obj
````

При вызове reindex для этого Series, данные будут реогранизованы в соответствии с новым индексом, а если каких--то из имеющихся в индексе значений не было, вместо них будут подставлены NA:

````python
obj2 = obj.reindex(['a', 'b', 'c', 'd', 'e'])
obj2
````

В случае DataFrame функция ```reindex``` может изменять столбцы, строки или и то и другое.

Если ей передать просто последовательность, то в результирующем объекте переиндексируются строки:

````python
example = pd.DataFrame(np.arange(9).reshape((3, 3)),
                     index=['a', 'c', 'd'],
                     columns=['Ohio', 'Texas', 'California'])
example
````

````python
example = example.reindex(['a', 'b', 'c', 'd'])
example
````

Для переиндексации столбцов нужно задать ключевое слово columns:

````python
states = ['Texas', 'Utah', 'California']
example.reindex(columns=states)
````

### Arithmetic and Data Alignment

Одна из самых важных черт Pandas - поведение арифметических операций для объектов с разными индексами. Если при сложении двух объектов обнаруживаются различные пары индексов, то **результирующий индекс будет объединением индексов**.

````python
s1 = pd.Series([7.3, -2.5, 3.4, 1.5], index=['a', 'c', 'd', 'e'])
s2 = pd.Series([-2.1, 3.6, -1.5, 4, 3.1],
               index=['a', 'c', 'e', 'f', 'g'])
````

````python
s1
````

````python
s2
````

Сложение этих объектов даст:

````python
s1 + s2
````

Вследствие внутреннего выравнивания данных образуются отсутствующие значения в позициях, для которых **не нашлось соответствующей пары**. Эти отсутствующие значения распространяются на последующие операции.

````python
df1 = pd.DataFrame({'A': [1, 2]})
df2 = pd.DataFrame({'B': [3, 4]})
df1
````

````python
df2
````

````python
df1 - df2
````

#### Arithmetic methods with fill values
При выполнении арифметических действий с объектами проиндексированными по разному иногда желательно поместить специальное значение (например 0), в позиции операнда, которым в другом операнде соответствует отсутствующая позиция:

````python
df1 = pd.DataFrame(np.arange(12.).reshape((3, 4)),
                   columns=list('abcd'))
df2 = pd.DataFrame(np.arange(20.).reshape((4, 5)),
                   columns=list('abcde'))
````

Сложение этих объектов порождает отсутствующие значения в позициях, которые имеются не в обоих операндах:

При вызове  метода ```add``` объекта df1 можно передать объект df2 и значение параметра ```fill_value```:

#### Operations between DataFrame and Series

Арифметические операции между DataFrame и Series конкретно определены. Рассмотрим вычисление разности между матрицей и одной из ее строк:

````python
arr = np.arange(12.).reshape((3, 4))
arr
````

````python
frame - series
````

Если какой-либо индекс не найдется либо в столбцах DataFrame, либо в индексе Series, то объекты переиндексируются для образования объединения:

Если вы хотите сопостовлять строки, а укладывать столбцы, то должны будете воспользоваться каким-нибудь арифметическим методом.

Например:

````python
series3 = frame['d']
frame
````

````python
series3
````

````python
frame.sub(series3, axis=0)
````

### Function Application and Mapping

Универсальные функции NumPy (поэлементные методы массивов) отлично работают с объектами Pandas:

````python
frame = pd.DataFrame(np.random.randn(4, 3), columns=list('abc'),
                     index=['Utah', 'Ohio', 'Texas', 'Oregon'])
frame
````

````python
frame = np.abs(frame)
frame
````

Еще одна часто встречающаяся операция - применение функции, определенной для одномерных массивов к каждой строке или столбцу. Для этого применяется метод ```apply``` объекта DataFrame:

````python
def f(x):
    return x.max() - x.min()
````

````python
frame.apply(f)
````

Ту же функцию можно определить lambda функцией:

````python
f = lambda x: x.max() - x.min()
frame.apply(f)
````

Теперь применим ту же функцию, но уже по отношению к столбцам:

````python
frame.apply(f, axis=1)
````

**По умолчанию, операции применяются по отношению к строкам, для того, чтобы операции применялись по отношению к столбцам, необходимо явно задавать параметр ```axis=1```**

Многие из наиболее распространенных статистик массивов (например, ```sum``` и ```mean```) - методы DataFrame, поэтому применяеть apply в этом случае необязательно.

````python
frame.sum(axis=1)
````

Функция передаемая объекту apply не обязана возвращать скалярное значение, она может вернуть и объект Series, содержащий несколько значений:

````python
def f(x):
    return pd.Series([x.min(), x.max()], index=['min', 'max'])
frame.apply(f)
````

Можно использовать и поэлементные функции Python. Допустим, требуется перевести все дробные значения в целые числа. Это позволяет сделать метод ```applymap```:

````python
format = lambda x: int(x)
frame.applymap(format)
# frame.map(format)  # applymap упраздняется, используем просто map
````

Этот метод называется ```applymap``` поскольку в классе Series уже есть метод ```map``` для применения функции к каждому элементу:

````python
frame['b'].map(format) # теперь также работает и для DataFrame
````

Для замены значений на уровне Series или DataFrame существует метод replace:

````python
frame['b'].map(format).replace(0, 100)
````

````python
frame['a'].map(format)
````

````python
frame['a'].map(format).replace([1, 0], [10, 100]) # можно использовать списки
````

````python
frame.map(format).replace({1: "здесь была 1", 0: "здесь был 0"}) # а можно словарь
````

### Unique Values, Value Counts, and Membership

Еще один класс методов служит для извелечения информации о значениях, хранящихся в одномерном объекте Series. Для иллюстрации рассмотрим пример:

````python
obj = pd.Series(['c', 'a', 'd', 'a', 'a', 'b', 'b', 'c', 'c'])
obj
````

Метод ```unique``` возвращает массив уникальных значений в Series:

````python
uniques = obj.unique()
uniques
````

````python
uniques = np.array(list(set(obj))).astype('object') # аналогичная операция, но без unique
uniques
````

Уникальные значения необязательно возвращаются в отсортированном порядке, но могут быть отсортированы впоследствии если это необходимо (```np.sort()```). Метод ```value_counts()``` вычисляет объект Series, содержащий частоты встречаемости значений:

````python
obj.value_counts() # 100% пригодится для анализа категориальных признаков
````

Функция ```value_counts``` может быть вызывана как метод, и в таком случае применима к **любому массиву или последовательности**

````python
pd.value_counts(obj.values, sort=False)
````

Метод ```isin``` вычисляет булев вектор членства в множестве и полезен для фильтрации набора данных относительно подмножества значений в объекте Series или столбце DataFrame:

````python
obj
````

````python
mask = obj.isin(['b', 'c'])
mask
````

````python
obj[mask] # убери все, кроме c и b
````

````python
obj[~mask] # оставь все, кроме c и b
````

Итак,

* ```isin``` вычисляет булев массив, показывающий, содержится ли каждое принадлежащее Series значение в переданной последовательности;
* ```unique``` вычисляет массив уникальных значений в Series и возвращает их в порядке появления
* ```value_counts``` возвращает объект Series который содержит уникальное значение (как индекс) и его частоту (как значение). Отсортирован в порядке убывания частот.
